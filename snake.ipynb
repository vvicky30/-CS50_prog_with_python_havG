```python
# program to illustrate the usage of generators in python , specifically in the scenerio where we have to generate large numbers of things (in terms of millions and billions).

def main():
    num= int(input("how many snakes do you want?"))
    #let's now print out each snake flocks at a time [will be increamneted style]
    for snakes in snake_flocks(num):
        print(snakes)  

def snake_flocks(n):
    # first we initialized flock list here where it will save the increamenting  flocks of snakes
    flock = []
    for i in range(n):
        flock.append("🐍" * i)
    return flock  # this will return the flocks of snakes in increamented order in one go 
                   # that further will be printed by main drive code function using print
                   
if __name__=="__main__":
    main()
```

```python
# if we use same code to print 10000 such flocks of snakes ? can we ? [obviously it takes some time]
def main():
    num= int(input("how many snakes do you want?"))
    #let's now print out each snake flocks at a time [will be increamneted style]
    for snakes in snake_flocks(num):
        print(snakes)  

def snake_flocks(n):
    # first we initialized flock list here where it will save the increamenting  flocks of snakes
    flock = []
    for i in range(n):
        flock.append("🐍" * i)
    return flock  # this will return the flocks of snakes in increamented order in one go 
                   # that further will be printed by main drive code function using print
                   
if __name__=="__main__":
    main()
```

```python
# if we tries to run above code for genrating  million flocks of snakes ,,then whole IDE will be crashed while executing the program.

# this is because our snake_flocks fn is gneerating flocks of snakes n-times by increamenting them +i-times on each line 
# this fn is generating all these flocks in one go and passing the flock list full of those generated flocks to print , 
# which's quite heavy to print even quite heavy to generate for function as well in one go 

# so solution for this is we will use Yield instead of return function(which returns list full of generate flocks) due to this function will generate each flock one at a time and pass one at a time to main function for printing 

def main():
    num= int(input("how many snakes do you want?"))
    #let's now print out each snake flocks at a time [will be increamneted style]
    for snakes in snake_flocks(num):
        print(snakes)  

def snake_flocks(n):
    for i in range(n):
        yield "🐍" *i  # this will return flocks of snakes (i -times) at a time , one at a time to main driver code function to print
    
                   
if __name__=="__main__":
    main() 
    
```
