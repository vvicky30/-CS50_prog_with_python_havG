### Full Pyramid Patterns in Python
A full pyramid pattern is a series of lines that form a pyramid-like structure. 

Each line contains a specific number of characters, and the number of characters on each line increases symmetrically as we move down the pyramid.

```python
# for supppose we have to make pyramid of hieght 5
#first loop for iterating over rows of pyramid ; here this loop iterate over 5 rows total as height =5
#second loop for iterating column-wise for each row , where this one specifically print leading spaces 
#third loop will print the symbol after leading spaces and for each succesive row (or for each first loop) this lopp print 2 more symbol wrt last row
 
def main1():
    star_pyramid(6)
    print("------------------------")
    star_py(5) # here while defining this fn we're using short approach , where we combine second loop (with purpose of printing leading spaces column-wise for aech row)

def star_pyramid(n=0): # passing default hieght of pyramid as zero , or default rows of pyramid as zero 
    for i in range(1 , n + 1):
        
        for j in range(n - i):
            print(" ", end= "")
            
        for k in range(1, 2*i):
            print("*", end="")
    
        print() # for changing line for each succesive row after printing leading spaces[and symbols after leading spaces] 
           
def star_py(n=0):
    for i in range(1, n+1):
        print(" " *(n-i), end="")
        
        for k in range(1, 2*i):
            print("*", end="" )
        
        print()        

main1()
```

```python
# now if we want to print pyramid pattern for numbers 
def main4():
    full_num_pyr(6)
    print("------------------------")
    half_num_pyr(6)
    print("------------------------")
    # what if we want to print full and half invert pyramid 
    # for that we have to just invert the sequqnces of those for loops 
    full_num_pyr_inv(6)
    print("------------------------")
    half_num_pyr_inv(6)
    
    print("^V^V^V^V^V^V^V^V^V^V^V^V^V^V^V^V^V^V")
    num_pyr_diamond(8) # if we have to print diamonds of numbers then basically we have to combine full_num_pyramid and full_num_inverted_pyramid
    
    
    
def full_num_pyr(n=0):
    for i in range(1 , n+1):
        print(" " * (n-i), end="")
        for k in range(1,2*i):
            print(k, end="")
        
        print()
    
def half_num_pyr(n=0):
    for i in range(1 , n+1):
    
        for k in range(1,2*i):
            print(k, end="")
        
        print()


def full_num_pyr_inv(n=0):
    for i in range(n, 0, -1):
        
        for j in range(n-i):
            
            print(" ", end="") # this printing fns for leading spaces now from bottom to top
        
        for k in range(2*i-1):
            print(k+1, end="")
        
        print("\r")# instead of new line here we have use "carriage return(\r)" with print to move cursor to the begninning of the current line
        
    
def half_num_pyr_inv(n=0):
    for i in range(n,0,-1):
        
    
        for k in range(2*i-1):
            print(k+1, end="")
        
        print("\r")# instead of new line here we have use "carriage return(\r)" with print to move cursor to the begninning of the current line
    

# here we combine full regular and inverted number pyramid  , basically called full regular and full inverted num pyramid 
def num_pyr_diamond(n=0):   
    # as we know here n is height of overall diamond 
    # while calling each pyramid fn we have to half them up their respective hieghts wrt diamond's hieghts , so that after combining the two pyramid of half-half hieght will resulted in diamond of required height     
   m = int(n/2)
   return(full_num_pyr(m),full_num_pyr_inv(m))
   
   
main4()
   
```

```python
def hollow_pyramid(n):
    for i in range(1, n + 1):
        for j in range(1, 2 * n):
            if i==n or j-i == n - 1 or j+i == n + 1:
                print("*", end="")
            else:
                print(" ", end="")
        print()
 
# Set the number of rows for the pyramid
rows = 10
 
# Call the function with the specified number of rows
hollow_pyramid(rows)
```

```python
def hollow_pyramid_neg(n): # here we print negative image of hollow pyramid for that we just have to print spaces first based on condition 
    for i in range(1, n + 1):
        for j in range(1, 2 * n):
            if i==n or j-i == n - 1 or j+i == n + 1:
                print(" ", end="")   # printing spaces when these three conditions are ture 
            else:
                print("*", end="")  # printing symbols elsewhere, 
        print() # new line 
 
# Set the number of rows for the pyramid
rows = 10
 
# Call the function with the specified number of rows
hollow_pyramid_neg(rows)
```

```python
def hollow_T(n): # printing  'T' that also inverted thrgough changing minor conditions of same above code of printing hollow pyramid
    for i in range(1, n + 1):
        for j in range(1, 2 * n):
            if i==n or j-i == n - i or j+i == n + i:   # here in place of n-1 and n+1 we used n-i and n+i
                print("*", end="")
            else:
                print(" ", end="")
        print()
 
# Set the number of rows for the pyramid
rows = 10
 
# Call the function with the specified number of rows
hollow_T(rows)
```
