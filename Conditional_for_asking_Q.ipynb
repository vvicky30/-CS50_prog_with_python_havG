```python
'''
Python if else statement in a List Comprehension
In this example, we are using an if statement in a list comprehension with 
the condition that if the element of the list is odd 
then its digit sum will be stored else not.
'''

def odd_digi_sum(n):
    Od_dg_sum = 0 # counter initialisation
    for ele in str(n):
        Od_dg_sum += int(ele)
    return(Od_dg_sum)

num_list = [236, 455, 788, 194, 193, 123, 122, 636]
print("this is given number list: ", num_list)
#now we use list comphresion method here
odd_sum = [odd_digi_sum(i) for i in num_list if i&1 ] 
print("here is the list of sums of only odd elements in given number of list: ", odd_sum)
#here as you can see with the help of list comphrension method we're passing only those elements from the list which's odd[if i&1] and only sum of those odd numbers will be stored in another list
```
