```python
'''
Python if else statement in a List Comprehension
In this example, we are using an if statement in a list comprehension with 
the condition that if the element of the list is odd 
then its digit sum will be stored else not.
'''

def odd_digi_sum(n):
    Od_dg_sum = 0 # counter initialisation
    for ele in str(n):
        Od_dg_sum += int(ele)
    return(Od_dg_sum)

num_list = [236, 455, 788, 194, 193, 123, 122, 636]
print("this is given number list: ", num_list)
#now we use list comphresion method here
odd_sum = [odd_digi_sum(i) for i in num_list if i&1 ] 
print("here is the list of sums of only odd elements in given number of list: ", odd_sum)
#here as you can see with the help of list comphrension method we're passing only those elements from the list which's odd[if i&1] and only sum of those odd numbers will be stored in another list
```

### Nested-If Statement in Python

 Python allows us to nest if statements within if statements. i.e., we can place an if statement inside another if statement.


Syntax: 

    if (condition1):
    
           Executes when condition1 is true
   
             if (condition2): 
       
                     Executes when condition2 is true 
                     [provided that condition-1 already satisfied]

## Python if-elif-else Ladder
### Here, a user can decide among multiple options. The if statements are executed from the top down.

As soon as one of the conditions controlling the if is true, the statement associated with that if is executed, and the rest of the ladder is bypassed. If none of the conditions is true, then the final else statement will be executed.
