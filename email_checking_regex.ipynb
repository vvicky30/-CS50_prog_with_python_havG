### Syntax for regex:-

### Metacharacters
#### Metacharacters are the characters with special meaning.

- To understand the RE analogy, Metacharacters are useful and important. They will be used in functions of module re. Below is the list of metacharacters.

\

Used to drop the special meaning of character following it

[]

Represent a character class

^

Matches the beginning

$

Matches the end

.

Matches any character except newline

|

Means OR (Matches with any of the characters separated by it.

?

Matches zero or one occurrence

*

Any number of occurrences (including 0 occurrences)

+

One or more occurrences

{}

Indicate the number of occurrences of a preceding regex to match.

()

Enclose a group of Regex



```python
#Improvisation- 1st
Email_1 = input("Enter your Email").strip()
if re.search(".*@.*", Email_1):  # here used dot means any character except for new line (but not a one time)
    # there might be the case or one or more chracter before and after @ ...so here we use ".*" before and after literally "@"
    # hefre * used for zero or more repitions of dot(and here dot means any chracter except newline)
    
    print("Email is valid")
else:
    print("Email is Invalid")
    
'''
o/p:(cutting-test cases for above code):-
-bahika@    //mail without domain considered here valid , because * in regex denotes to zero or more repitions instead of 1 or more repetitions 
>Email is valid
'''
```

#### As 're'(regex) is the official liberary to check the any string against the specified regular-expression 
- behind the scenes this liberary used deidcated matching-machines software , in formal terms we call them finite-state-machines or Non-Deterministic Finite Automata

- here below is the simple finite automata diagram of above regualr expression : " .* @ .* "
  



### explanation:
- here q1(single circled) is initial state
- here q2(double circled) is final state 
- q1 self looped with any character (except newline) as input[* times : means zero or more times] 
- once it encountered with "@" it will change the state from q1 to q2(which is final state as well) 
- now q2 self looped with any character (except newline) as input[* times : means zero or more times] // means without taking any chracter after '@'it will considered ending of machine-state   


```python
#Improvisation- 2nd
Email_2 = input("Enter your Email").strip()
if re.search(".+@.+", Email_2):  # as here we used + in the place of * , and here + denotes to one or more repetations (not zero or more)
    print("Email is valid")
else:
    print("Email is Invalid")
    
'''
o/p:(cutting-test cases for above code):-
-bahika@    //mail without domain
>Email is Invalid

-bahika@notdotcom //in domain name there's not '.com' mentioned still considered to be valid
>Email is valid
'''
```

```python
#Improvisation- 3rd
Email_3 = input("Enter your Email").strip()

#if re.search(".+@.+.com", Email_3):  # as here we used add '.com' in the last , to make sure that user should enter .com while ending domain section
# but here little bit problem , as here dot of '.com' will be interpreted as metacharacter-dot by regex machine 
# for solving this problem we can use backslash '\' which is escape-symbol used here to escape of metacharacter interpretation by regex
# so we going to use "\.com" instead of ".com" here :

if re.search(r".+@.+\.com", Email_3):    # here 'r' before wtiring regular expression used to denotes whatever write (or string after 'r') 
    #should be considered as raw string.
    
    print("Email is valid")
else:
    print("Email is Invalid")
    
'''
o/p:(cutting-test cases for above code):-

-bahika@notdotcom //in domain name there's not '.com' mentioned now considered to be Invalid
>Email is Invalid

but now:
-bahika@nowdot.com
>Email is valid   so code working fine 

-bahika@nowdot?com
>Email is Invalid
'''
```

#### note: The raw string (r"---")is slightly different from a regular string, it won’t interpret the \ character as an escape character. This is because the regular expression engine uses \ character for its own escaping purpose.

```python
#Improvisation- 4th
Email_4 = input("Enter your Email").strip()

if re.search(r"^.+@.+\.com$", Email_4): # here we used ^(carrot) to mark start of string with following expression.
                                       # whereas we also use $(dollar sign) to make end of the expression with trailing chracters [here it's  '.com'.
    print("Email is valid")
else: 
    print("Email is Invalid")
    
'''
o/p:(cutting-test cases for above code):-
-here's is my email vahika@gmail.com    //as its not starting with mail expression 
>Email is Invalid


-vahika@gmail.com
>Email is valid

-vahika@@@@gmail.com   //let's imrovise for this test case
>Email is valid
'''
```

### for improvising above test case with multiple '@' symbol in our email, and still considered to be valid email above:
- 1. '[]' this square bracket used for assigning multiple chracters, sets of characters , range of sets for characters to be included etc.

- 2. '[^]' where this square bracket with carrot is complementing the specified set in the bracket 


```python
#Improvisation- 5th
Email_5 = input("Enter your Email").strip()

#here we are going to restrict the chracters before and after '@' to be anything except '@' itself
#here we have to replace '.' before after '@' with '[^@]'
# here '[^@]' replacement means , while scanning....before and after literal one '@' there should be all character except for the '@'.
# so in this way we're restricitng the mail-test-cases with multiple '@' symbol 
if re.search(r"^[^@]+@[^@]+\.com$", Email_5):
     print("Email is valid")
else: 
    print("Email is Invalid")

'''
o/p:(cutting-test cases for above code):-
-vahika@@@@gmail.com
>Email is Invalid
'''
```

- in the above regular expression which is : r"^[^@]+@[^@]+\.com$"

-> instead of complementing the '@' symbol through '[^@]' before 
after @, we can specifically allowed uppercase alphabets, 
lowercase alphabets alongwith 0 to 9 numbers and underscores in 
username section(before @) and in domain section(after @)
we can do that like this : '[a-zA-Z0-9_]'
- r"^[a-zA-Z0-9_]+@[a-zA-Z0-9_]+\.com$" 

-> even above expression can be shortened further with replacing [a-zA-Z0-9_] with '\w' like this :

- r"^\w+@\w+\.com$"
 


```python
#Improvisation- 6th
Email_6 = input("Enter your Email").strip()
if re.search(r"^\w+@\w+\.com$", Email_6):
     print("Email is valid")
else: 
    print("Email is Invalid")

'''
o/p:(cutting-test cases for above code):-
-videshi_gautam1930@gmail.com
>Email is valid

it's not nessacary that domain section ends with com but it can be ends with edu or org or gov
like this test case:
-videshi_gautam1930@gmail.gov  //even domain ends with .gov here it's still considered to be invalid email
>Email is Invalid
 let's improvise for all these several domain-names endings or extension
'''
```

```python
# we can make or group all those global domain extension to be allowed for considering valid mail 
# we can group them in round brackets parentheses like this :- (com|edu|gov|org)
#Improvisation- 7th
Email_7 = input("Enter your Email").strip()
if re.search(r"^\w+@\w+\.(com|edu|org|gov)$", Email_7):
     print("Email is valid")
else: 
    print("Email is Invalid")

'''
o/p:(cutting-test cases for above code):-
-videshi_gautam1930@gmail.gov
>Email is valid
-VIDESHI_1930@GMAIL.ORG     //here we write our whole email with caps, generally Emails are case-insensitive but again here it's considered as invalid
>Email is Invalid

so now again we're going improvise our regular expression in the way that it will ignore cases of letters in the email
'''
```

### so now again we're going improvise our regular expression in the way that it will ignore cases of letters in the email

### as this is the regex syntax we know already (where we not used 'flag' the third argument here yet in the code):-



- in this flags argument we can used all of this 
1. re.IGNORECASE
2. re.MULTILINE
3. re.DOTALL

- specifically here in our case we're going to use 1st one for making our regex case insensitive.

```python
#Improvisation- 8th
Email_8 = input("Enter your Email").strip()
if re.search(r"^\w+@\w+\.(com|edu|org|gov)$", Email_8, re.IGNORECASE):
     print("Email is valid")
else: 
    print("Email is Invalid")

'''
o/p:(cutting-test cases for above code):-

-VIDESHI_1930@GMAIL.ORG 
>Email is valid

but if we enter the email like this which has to be valid , but still considered here invalid by this code
-videshi_19@cs50.gmail.com
>Email is Invalid
next for this test case we're going to improvise 
'''
```

```python
# here this is 'cs50.' like subsection of whole domain section can be called as subdomain-section 
# this subdomain section usually entered by user optionally not necessarily all the time 
# so we're going to make this subdomain section as optional one as : either 'cs50.' or none
#'cs50.' --> regex: (\w+\.)?     whereas ? used for zero or one repetation meaning optional.
#Improvisation- 9th
Email_9 = input("Enter your Email").strip()
if re.search(r"^\w+@(\w+\.)?\w+\.(com|edu|org|gov)$", Email_9, re.IGNORECASE):
     print("Email is valid")
else: 
    print("Email is Invalid")

'''
o/p:(cutting-test cases for above code):-

-videshi_19@cs50.gmail.com 
>Email is valid

'''
```
